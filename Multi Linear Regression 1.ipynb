Convert our variables datatype from series to array 

```python
x1 = np.array(x1)
x2 = np.array(x2)
x3 = np.array(x3)
x4 = np.array(x4)
x5 = np.array(x5)
x6 = np.array(x6)
x7 = np.array(x7)
x8 = np.array(x8)

y = np.array(y)
```

```python
# x1
```

Number of rows

```python
n = len(x1)

print(n)
```

 Create a â€œonesâ€ matrix 

```python
x_bias = np.ones((n , 1))
```

```python
x_bias
```

```python
x_bias.shape
```

 Reshape our data so that we can perform operations like addition and multiplication with x_bias 

```python
x1_new = np.reshape(x1 , (n , 1))
x2_new = np.reshape(x2 , (n , 1))
x3_new = np.reshape(x3 , (n , 1))
x4_new = np.reshape(x4 , (n , 1))
x5_new = np.reshape(x5 , (n , 1))
x6_new = np.reshape(x6 , (n , 1))
x7_new = np.reshape(x7 , (n , 1))
x8_new = np.reshape(x8 , (n , 1))
```

```python
x1_new
```

Create a major matrix with all the columns like x_bias , x1_new , ........ , x8_new

```python
x_new = np.append(x_bias , x1_new , axis = 1)
x_new = np.append(x_new , x2_new , axis = 1)
x_new = np.append(x_new , x3_new , axis = 1)
x_new = np.append(x_new , x4_new , axis = 1)
x_new = np.append(x_new , x5_new , axis = 1)
x_new = np.append(x_new , x6_new , axis = 1)
x_new = np.append(x_new , x7_new , axis = 1)
x_new = np.append(x_new , x8_new , axis = 1)
```

MAJOR MATRIX

```python
print(x_new)
```

CALCULATION

Find transpose of a matrix

```python
x_new_transpose = np.transpose(x_new)
```

Perform multiplication (dot product of original and transpose matrix)

```python
x_new_transpose_dot_x_new = x_new_transpose.dot(x_new)
```

Find inverse

```python
temp_1 = np.linalg.inv(x_new_transpose_dot_x_new)
```

perform multiplication (dot product of transposed x and y)

```python
temp_2 = x_new_transpose.dot(y)
```

FINDING COFFECIENTS

```python
theta = temp_1.dot(temp_2)
```

SLOPE AND COEFFICIENTS VALUES

```python
print(theta)
```

```python
beta_0 = theta[0]
beta_1 = theta[1]
beta_2 = theta[2]
beta_3 = theta[3]
beta_4 = theta[4]
beta_5 = theta[5]
beta_6 = theta[6]
beta_7 = theta[7]
beta_8 = theta[8]
```

PREDICTION FUNCTION

```python
def predict(beta_0,beta_1,beta_2,beta_3,beta_4,beta_5,beta_6,beta_7,beta_8 , f1,f2,f3,f4,f5,f6,f7,f8) :
    
    predicted_Value = beta_0 + f1*beta_1 + f2*beta_2 + f3*beta_3 + f4*beta_4 + f5*beta_5 + f6*beta_6 + f7*beta_7 + f8*beta_8
    
    return predicted_Value
```
